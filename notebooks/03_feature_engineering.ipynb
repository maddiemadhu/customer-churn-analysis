## Import Libraries

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

## Load Cleaned Dataset

```python
df = pd.read_csv("../data/processed/clean_telco_churn.csv")

df.head()
```

## Check Data Types

```python
df.info()
```

## Reconfirm Numeric Columns

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

## Drop Customer ID (Not Useful for Modeling)

```python
df = df.drop("customerID", axis=1)
```

## Encode Categorical Variables

```python
df = pd.get_dummies(df, drop_first=True)
```

## Separate Features & Target

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

## Feature Scaling (Important for Logistic Regression)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

#  Feature Engineering — Key Insights

* Converted categorical variables into numeric using One-Hot Encoding
* Used `drop_first=True` to avoid multicollinearity
* Removed `customerID` as it has no predictive value
* Converted `TotalCharges` to numeric format
* Split data into features (X) and target (y)
* Applied StandardScaler for feature scaling
* Ensured all features are on the same scale for better model performance
